# Functions

## Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate an array of values from -10 to 10
x_values = np.linspace(-10, 10, 400)
y_values = sigmoid(x_values)

plt.figure(figsize=(10, 6))

# Plot the sigmoid function
plt.plot(x_values, y_values, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

## ReLU (Rectified Linear Unit)

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

# Generate an array of values from -10 to 10
x_values = np.linspace(-10, 10, 400)
y_values = relu(x_values)

plt.figure(figsize=(10, 6))

# Plot the ReLU function
plt.plot(x_values, y_values, label='ReLU Function')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

## All Activation Functions

There are a variety of activation functions used in neural networks, each with its own set of advantages and disadvantages. Here's a rundown of some of the more commonly used ones:

### Linear
- **Equation**: \( f(x) = x \)
- **Pros**: Simplicity
- **Cons**: Cannot model complex, nonlinear relationships

### Sigmoid
- **Equation**: \( f(x) = \frac{1}{1 + e^{-x}} \)
- **Pros**: Maps output between 0 and 1
- **Cons**: Vanishing gradient problem, not zero-centered

### Hyperbolic Tangent (tanh)
- **Equation**: \( f(x) = \tanh(x) \)
- **Pros**: Output range between -1 and 1, zero-centered
- **Cons**: Vanishing gradient problem

### Rectified Linear Unit (ReLU)
- **Equation**: \( f(x) = \max(0, x) \)
- **Pros**: Computationally efficient, mitigates vanishing gradient
- **Cons**: Dying ReLU problem (neurons can sometimes get "stuck" during training)

### Leaky ReLU
- **Equation**: \( f(x) = x \) if \( x > 0 \), \( f(x) = \alpha x \) otherwise (\( \alpha \) is a small constant)
- **Pros**: Addresses the dying ReLU problem
- **Cons**: Not zero-centered

### Parametric ReLU (PReLU)
- **Equation**: Similar to Leaky ReLU, but \( \alpha \) is learned during training
- **Pros**: More flexible than Leaky ReLU
- **Cons**: Risk of overfitting on small datasets

### Exponential Linear Unit (ELU)
- **Equation**: \( f(x) = x \) if \( x > 0 \), \( f(x) = \alpha (e^x - 1) \) otherwise
- **Pros**: Addresses dying ReLU problem, speeds up learning
- **Cons**: Computationally expensive

### Softmax
- **Equation**: \( f(x)_i = \frac{e^{x_i}}{\sum_{j} e^{x_j}} \) for each element \( i \) in vector \( x \)
- **Pros**: Useful for multi-class classification, converts scores to probabilities
- **Cons**: Not suitable for non-classification tasks

### Swish
- **Equation**: \( f(x) = x \cdot \sigma(\beta x) \), \( \sigma \) is the sigmoid function, \( \beta \) is a learnable parameter
- **Pros**: Smooth curve, helps with gradient flow
- **Cons**: Computationally more expensive than ReLU

Each of these activation functions has specific use-cases, and the choice often depends on the specific problem you're trying to solve. Researchers are continually coming up with new activation functions to address specific challenges in training neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0, keepdims=True)

def swish(x, beta=1):
    return x * sigmoid(beta * x)

# Generate an array of values from -10 to 10 for plotting
x_values = np.linspace(-10, 10, 400)

# Create subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# Linear
axs[0, 0].plot(x_values, linear(x_values))
axs[0, 0].set_title('Linear')

# Sigmoid
axs[0, 1].plot(x_values, sigmoid(x_values))
axs[0, 1].set_title('Sigmoid')

# Tanh
axs[0, 2].plot(x_values, tanh(x_values))
axs[0, 2].set_title('Tanh')

# ReLU
axs[1, 0].plot(x_values, relu(x_values))
axs[1, 0].set_title('ReLU')

# Leaky ReLU
axs[1, 1].plot(x_values, leaky_relu(x_values))
axs[1, 1].set_title('Leaky ReLU')

# ELU
axs[1, 2].plot(x_values, elu(x_values))
axs[1, 2].set_title('ELU')

# Softmax
axs[2, 0].plot(x_values, softmax(x_values))
axs[2, 0].set_title('Softmax')

# Swish
axs[2, 1].plot(x_values, swish(x_values))
axs[2, 1].set_title('Swish')

# Empty plot to balance grid
axs[2, 2].axis('off')

# Add grid to all subplots
for ax in axs.flat:
    ax.grid(True)

plt.show()
```
