# Dataset - Diabetes

## Fundamentals

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

# Load diabetes dataset from sklearn
diabetes = load_diabetes()

# Convert dataset into pandas DataFrame
diabetes_df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
diabetes_df['target'] = diabetes.target

diabetes_df.describe()
```

## Scatter

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select subset of features for visualization
subset_features = ['bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6', 'target']

# Create a pairplot (scatter plot matrix)
sns.pairplot(diabetes_df[subset_features])

# Show the plot
plt.show()
```

## Distributions with Histograms

```python
# Create a figure with multiple subplots
fig, axs = plt.subplots(3, 4, figsize=(20, 15))

# Create histograms for each feature
features = diabetes.feature_names + ['target']
for i, ax in enumerate(axs.flat):
    if i < len(features):
        ax.hist(diabetes_df[features[i]], bins=30, color='skyblue', edgecolor='black')
        ax.set_title(features[i])

# Remove empty subplots
for i in range(len(features), len(axs.flat)):
    fig.delaxes(axs.flat[i])

# Show the plot
plt.tight_layout()
plt.show()
```

## Use Meaninful Names

```python
# Change the column names to the detailed attribute names
diabetes_df.columns = ['Age (years)', 'Sex', 'Body mass index', 'Average blood pressure', 
                       'Total serum cholesterol', 'Low-density lipoproteins',
                       'High-density lipoproteins', 'Total cholesterol / HDL',
                       'Log of serum triglycerides level', 'Blood sugar level', 'Target']

# Create a figure with multiple subplots
fig, axs = plt.subplots(4, 3, figsize=(20, 20))

# Create histograms for each feature
features = diabetes_df.columns
for i, ax in enumerate(axs.flat):
    if i < len(features):
        ax.hist(diabetes_df[features[i]], bins=30, color='lightgreen', edgecolor='black')
        ax.set_title(features[i], fontsize=12)
        plt.setp(ax.get_xticklabels(), fontsize=10)
        plt.setp(ax.get_yticklabels(), fontsize=10)

# Remove empty subplots
for i in range(len(features), len(axs.flat)):
    fig.delaxes(axs.flat[i])

# Show the plot
plt.tight_layout()
plt.show()
```
