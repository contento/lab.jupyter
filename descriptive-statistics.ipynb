# Basic Statistics

## Numerical

- **Mean**: Average value
- **Variance**: Measure of data dispersion
- **Standard Deviation**: Square root of the variance
- **Covariance**: Measure of the joint variability of two random variables
- **Correlation**: Measures the strength and direction of the relationship between two variables
- **Mode**: The number that appears most frequently
- **Median**: The middle value in a dataset
- **Kurtosis**: Measure of "tailedness" of the distribution
- **Skewness**: Measure of the asymmetry of the data around the sample mean

```python
import numpy as np
from scipy import stats
import pandas as pd

# Create two example datasets
data1 = np.array([2, 4, 4, 4, 5, 5, 7, 9])
data2 = np.array([10, 6, 8, 6, 5, 8, 4, 5])

# Mean
mean1 = np.mean(data1)
mean2 = np.mean(data2)
print(f"Mean of data1: {mean1}, Mean of data2: {mean2}")

# Variance
var1 = np.var(data1)
var2 = np.var(data2)
print(f"Variance of data1: {var1}, Variance of data2: {var2}")

# Standard Deviation
std_dev1 = np.std(data1)
std_dev2 = np.std(data2)
print(f"Standard Deviation of data1: {std_dev1}, Standard Deviation of data2: {std_dev2}")

# Covariance
covariance = np.cov(data1, data2)[0, 1]
print(f"Covariance between data1 and data2: {covariance}")

# Correlation
correlation = np.corrcoef(data1, data2)[0, 1]
print(f"Correlation between data1 and data2: {correlation}")

# Mode
mode1 = stats.mode(data1)[0]
mode2 = stats.mode(data2)[0]
print(f"Mode of data1: {mode1}, Mode of data2: {mode2}")

# Median
median1 = np.median(data1)
median2 = np.median(data2)
print(f"Median of data1: {median1}, Median of data2: {median2}")

# Kurtosis
kurtosis1 = stats.kurtosis(data1)
kurtosis2 = stats.kurtosis(data2)
print(f"Kurtosis of data1: {kurtosis1}, Kurtosis of data2: {kurtosis2}")

# Skewness
skewness1 = stats.skew(data1)
skewness2 = stats.skew(data2)
print(f"Skewness of data1: {skewness1}, Skewness of data2: {skewness2}")

# Creating a DataFrame for better visualization
df = pd.DataFrame({
    'Statistical Measure': ['Mean', 'Variance', 'Standard Deviation', 'Covariance', 'Correlation', 'Mode', 'Median', 'Kurtosis', 'Skewness'],
    'Data1': [mean1, var1, std_dev1, covariance, correlation, mode1, median1, kurtosis1, skewness1],
    'Data2': [mean2, var2, std_dev2, None, None, mode2, median2, kurtosis2, skewness2]
})

print("\nSummary in DataFrame:")
print(df)
```

## Visually

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

# Create two example datasets
data1 = np.array([2, 4, 4, 4, 5, 5, 7, 9])
data2 = np.array([10, 6, 8, 6, 5, 8, 4, 5])

# Calculate statistics
mean1, mean2 = np.mean(data1), np.mean(data2)
var1, var2 = np.var(data1), np.var(data2)
std_dev1, std_dev2 = np.std(data1), np.std(data2)

# Plotting Histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.hist(data1, bins=5, alpha=0.7, label='data1')
plt.axvline(mean1, color='r', linestyle='dashed', linewidth=1, label=f"Mean {mean1}")
plt.title('Histogram of data1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(2, 2, 2)
plt.hist(data2, bins=5, alpha=0.7, label='data2')
plt.axvline(mean2, color='r', linestyle='dashed', linewidth=1, label=f"Mean {mean2}")
plt.title('Histogram of data2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Scatter Plot
plt.subplot(2, 2, 3)
plt.scatter(data1, data2)
plt.title('Scatter Plot between data1 and data2')
plt.xlabel('data1')
plt.ylabel('data2')

# Bar Graph for Mean, Variance, and Standard Deviation
plt.subplot(2, 2, 4)
labels = ['Mean', 'Variance', 'Std Dev']
data1_stats = [mean1, var1, std_dev1]
data2_stats = [mean2, var2, std_dev2]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, data1_stats, width, label='data1')
plt.bar(x + width/2, data2_stats, width, label='data2')
plt.xticks(x, labels)
plt.title('Mean, Variance, and Std Dev for data1 and data2')
plt.legend()

plt.tight_layout()
plt.show()
```

## Kurtosis and Skewness

```python
## Calculate Kurtosis and Skewness
kurtosis1, kurtosis2 = stats.kurtosis(data1), stats.kurtosis(data2)
skewness1, skewness2 = stats.skew(data1), stats.skew(data2)

# Create a DataFrame for better visualization
df = pd.DataFrame({
    'Statistical Measure': ['Mean', 'Variance', 'Standard Deviation', 'Kurtosis', 'Skewness'],
    'Data1': [mean1, var1, std_dev1, kurtosis1, skewness1],
    'Data2': [mean2, var2, std_dev2, kurtosis2, skewness2]
}) 
print("\nSummary in DataFrame:")
print(df)
```

## Skewness: Mean, Median, and Mode

```python
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt

# Generate the dataset
np.random.seed(0)
data = np.concatenate([np.random.normal(loc=0, scale=1, size=1000),
                       np.random.exponential(scale=2, size=1000),
                       np.random.uniform(low=-1, high=1, size=1000)])

# Calculate statistics
mean = np.mean(data)
mode = stats.mode(data)[0]
median = np.median(data)
quartiles = np.percentile(data, [25, 50, 75])

# Plotting the statistics
plt.figure(figsize=(10, 6))

# Histogram
plt.subplot(2, 1, 1)
plt.hist(data, bins=30, alpha=0.7)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label=f"Mean: {mean:.2f}")
plt.axvline(mode, color='g', linestyle='dashed', linewidth=1, label=f"Mode: {mode:.2f}")
plt.axvline(median, color='b', linestyle='dashed', linewidth=1, label=f"Median: {median:.2f}")
plt.title('Histogram of the Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Boxplot
plt.subplot(2, 1, 2)
plt.boxplot(data, vert=False)
plt.axvline(quartiles[0], color='r', linestyle='dashed', linewidth=1, label=f"Q1: {quartiles[0]:.2f}")
plt.axvline(median, color='b', linestyle='dashed', linewidth=1, label=f"Median: {median:.2f}")
plt.axvline(quartiles[2], color='g', linestyle='dashed', linewidth=1, label=f"Q3: {quartiles[2]:.2f}")
plt.title('Boxplot of the Dataset')
plt.xlabel('Value')
plt.ylabel('Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```
