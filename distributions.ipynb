# Distributions

## Simple histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset of 1000 random data points between 0 and 10
# np.random.seed(0)  # For reproducibility
data = np.random.uniform(0, 10, 1000)  # min=0, max=10, num_points=1000

# Create histogram
plt.hist(data, bins=20, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of a Randomly Generated Dataset')
plt.show()
```

## Distribution of Dice Rolls

```python
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
# np.random.seed(0)

# Determine a random number of dice between 1 and 10
num_dice = np.random.randint(1, 11)

# Initialize an array to store the sum of rolls
sum_rolls = np.zeros(1000, dtype=int)

# Simulate 1000 rolls for each die and sum them up
for _ in range(num_dice):
    rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=1000)
    sum_rolls += rolls

# Create the histogram
plt.hist(sum_rolls, bins=np.arange(num_dice-0.5, 6*num_dice+0.5, 1), edgecolor='black', rwidth=0.8)

# Add labels and title
plt.xlabel(f'Sum of {num_dice} Dice')
plt.ylabel('Frequency')
plt.title(f'Distribution of 1000 Rolls of {num_dice} Dice')

# Show the plot
plt.show()

print(f"Simulated rolling {num_dice} dice.")
```

## Common Area

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import gamma

# Parameters for the gamma distribution
shape = 2.0
scale = 1.0

# Generate random samples
samples = gamma.rvs(a=shape, scale=scale, size=1000)

# Calculate the PDF
x = np.linspace(0, 10, 1000)
pdf = gamma.pdf(x, a=shape, scale=scale)

# Plot the histogram and PDF
plt.hist(samples, bins=30, density=True, alpha=0.6, color='blue', label='Samples')
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

plt.xlabel('x')
plt.ylabel('Density')
plt.legend(loc='best')
plt.title(f'Gamma Distribution (={shape}, ={scale})')
plt.show()
```

## Binomial

```python
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Calculate the PMF (probability mass function)
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plot the PMF as a bar graph
plt.bar(x, pmf, width=0.8, color='blue', alpha=0.6, label='PMF')

plt.xlabel('Number of successes (x)')
plt.ylabel('Probability')
plt.legend(loc='best')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.show()
```

## Gauss

```python
from scipy.stats import norm
from scipy.stats import norm, kurtosis


# Parameters for the Gaussian distribution
mean = 0.0  # Mean (mu)
std_dev = 1.0  # Standard deviation (sigma)

# Calculate the PDF (probability density function)
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Plot the PDF
plt.plot(x, pdf, 'b-', lw=2, label='PDF')

plt.xlabel('x')
plt.ylabel('Density')
plt.legend(loc='best')
plt.title(f'Gaussian Distribution (={mean}, ={std_dev}, Kurtosis = {round(kurtosis(x), 2)})')
plt.grid(True)
plt.show()
```

## Poisson

```python
from scipy.stats import poisson

# Parameters for the Poisson distribution
lmbda = 4  # Average rate ()

# Calculate the PMF
x = np.arange(0, 3 * lmbda)
pmf = poisson.pmf(x, lmbda)

# Plot the PMF as a bar graph
plt.bar(x, pmf, width=0.8, color='blue', alpha=0.6, label='PMF')

plt.xlabel('Number of events (k)')
plt.ylabel('Probability')
plt.legend(loc='best')
plt.title(f'Poisson Distribution (={lmbda})')
plt.grid(True)
plt.show()
```

## Uniform

```python
a = 2
b = 6

# Continuous uniform distribution
x = np.linspace(a - 1, b + 1, 1000)
pdf = [1 / (b - a) if a <= xi <= b else 0 for xi in x]
plt.plot(x, pdf, 'b-', lw=2, label='Continuous Uniform')

# Discrete uniform distribution
x_discrete = np.arange(a, b + 1)
pmf = [1 / (b - a + 1) for _ in x_discrete]
plt.stem(x_discrete, pmf, 'r', markerfmt='ro', label='Discrete Uniform', basefmt=' ')

plt.xlabel('x')
plt.ylabel('Probability')
plt.legend(loc='best')
plt.title(f'Uniform Distribution (a={a}, b={b})')
plt.grid(True)
plt.show()
```

## Exponential

```python
from scipy.stats import expon

# Parameters for the exponential distribution
lmbda = 1.5  # Average rate ()

# Calculate the PDF
x = np.linspace(0, 4 * (1 / lmbda), 1000)
pdf = lmbda * np.exp(-lmbda * x)

# Plot the PDF
plt.plot(x, pdf, 'b-', lw=2, label='PDF')

plt.xlabel('Time between events (x)')
plt.ylabel('Density')
plt.legend(loc='best')
plt.title(f'Exponential Distribution (={lmbda})')
plt.grid(True)
plt.show()
```
