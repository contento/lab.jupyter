# Fourier

## FFT: Noise and Spectrum

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple signal with two frequencies
dt = 0.01  # sampling interval
Fs = 1 / dt  # sampling frequency
t = np.arange(0, 1, dt)

f_clean = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t) + np.sin(2*np.pi*200*t)  # sum of 3 frequencies
f = f_clean + 2.5*np.random.randn(len(t))  # add some noise

# Compute the Fast Fourier Transform (FFT)
n = len(t)
fhat = np.fft.fft(f, n)  # compute the FFT
PSD = fhat * np.conj(fhat) / n  # power spectrum (power per freq)
freq = (1/(dt*n)) * np.arange(n)  # create x-axis of frequencies in Hz
L = np.arange(1, np.floor(n/2), dtype='int')  # only plot the first half of freqs

# plotting the signal and its Fourier Transform
fig, axs = plt.subplots(3, 1)

plt.sca(axs[0])
plt.plot(t, f_clean, color='b', linewidth=2, label='Clean')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[1])
plt.plot(t, f, color='c', linewidth=1.5, label='Noisy')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[2])
plt.plot(freq[L], PSD[L], color='c', linewidth=2, label='Spectrum')
plt.xlim(freq[L[0]], freq[L[-1]])
plt.legend()

plt.show()
```

## Sine Wave

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple signal with a single frequency
dt = 0.01  # sampling interval
Fs = 1 / dt  # sampling frequency
t = np.arange(0, 2, dt)  # generate time vector for 1 second

# generate a signal
f_clean = np.sin(2*np.pi*5*t)  # single frequency 5Hz sine wave

# Compute the Fast Fourier Transform (FFT)
n = len(t)
fhat = np.fft.fft(f_clean, n)  # compute the FFT
PSD = fhat * np.conj(fhat) / n  # power spectrum (power per freq)
freq = (1/(dt*n)) * np.arange(n)  # create x-axis of frequencies in Hz
L = np.arange(1, np.floor(n/2), dtype='int')  # only plot the first half of freqs

# plotting the signal and its Fourier Transform
fig, axs = plt.subplots(2, 1)

plt.sca(axs[0])
plt.plot(t, f_clean, color='b', linewidth=2, label='Sine Wave')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[1])
plt.plot(freq[L], np.abs(PSD[L]), color='r', linewidth=2, label='Spectrum')  # plot the absolute value of the spectrum
plt.xlim(freq[L[0]], freq[L[-1]])
plt.legend()

plt.show()
```

## White Noise (Normal)

## Pink Noise

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_pink_noise(n_samples, alpha=1):
    """Generates pink noise using the IFFT and the power law for the power spectrum."""
    random_phase = np.exp(1j * (2 * np.pi * np.random.random(n_samples//2+1) - np.pi))
    s_scale = np.power(np.arange(1, n_samples//2+2), -alpha / 2.0)
    return np.fft.irfft(s_scale * random_phase, n=n_samples)

# Create a simple signal with a single frequency
dt = 0.01  # sampling interval
Fs = 1 / dt  # sampling frequency
t = np.arange(0, 1, dt)  # generate time vector for 1 second

# generate a signal
f_clean = np.sin(2*np.pi*5*t)  # single frequency 5Hz sine wave
f = f_clean + generate_pink_noise(len(t))  # add pink noise

# Compute the Fast Fourier Transform (FFT)
n = len(t)
fhat = np.fft.fft(f, n)  # compute the FFT
PSD = fhat * np.conj(fhat) / n  # power spectrum (power per freq)
freq = (1/(dt*n)) * np.arange(n)  # create x-axis of frequencies in Hz
L = np.arange(1, np.floor(n/2), dtype='int')  # only plot the first half of freqs

# plotting the signal and its Fourier Transform
fig, axs = plt.subplots(3, 1)

plt.sca(axs[0])
plt.plot(t, f_clean, color='b', linewidth=2, label='Clean')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[1])
plt.plot(t, f, color='c', linewidth=1.5, label='Noisy')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[2])
plt.plot(freq[L], np.abs(PSD[L]), color='r', linewidth=2, label='Spectrum')  # plot the absolute value of the spectrum
plt.xlim(freq[L[0]], freq[L[-1]])
plt.legend()

plt.show()
```

## Impulse Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple impulse function
dt = 0.01  # sampling interval
Fs = 1 / dt  # sampling frequency
t = np.arange(0, 1, dt)  # generate time vector for 1 second

# generate a signal
f_clean = np.zeros_like(t)
f_clean[len(t)//2] = 1  # single impulse at the center

# Compute the Fast Fourier Transform (FFT)
n = len(t)
fhat = np.fft.fft(f_clean, n)  # compute the FFT
PSD = fhat * np.conj(fhat) / n  # power spectrum (power per freq)
freq = (1/(dt*n)) * np.arange(n)  # create x-axis of frequencies in Hz
L = np.arange(1, np.floor(n/2), dtype='int')  # only plot the first half of freqs

# plotting the signal and its Fourier Transform
fig, axs = plt.subplots(2, 1)

plt.sca(axs[0])
plt.plot(t, f_clean, color='b', linewidth=2, label='Impulse')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[1])
plt.plot(freq[L], np.abs(PSD[L]), color='r', linewidth=2, label='Spectrum')  # plot the absolute value of the spectrum
plt.xlim(freq[L[0]], freq[L[-1]])
plt.legend()

plt.show()
```

## Constant Function


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
dt = 0.01  # sampling interval
t = np.arange(0, 10, dt)  # time vector

# Create a constant function
f = np.full(shape=t.shape, fill_value=1)

# Perform the Fourier Transform
fhat = np.fft.fft(f)

# Compute the frequency spectrum
freqs = np.fft.fftfreq(f.size, dt)

# Create two subplots: one for the function and one for its Fourier Transform
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# Plot the function
axs[0].plot(t, f, label='Constant function')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Amplitude')
axs[0].legend()

# Plot the absolute value of the FFT result
axs[1].plot(freqs, np.abs(fhat), label='FFT')
axs[1].set_xlabel('Frequency')
axs[1].set_ylabel('Magnitude')
axs[1].legend()

# Show the plots
plt.tight_layout()
plt.show()
```
