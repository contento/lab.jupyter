# Filters

## Spatial

### Medical - built-in

```python
from skimage import filters, data
import matplotlib.pyplot as plt

# Load built-in cell image
image = data.cell()

# Apply Gaussian blur filter (smoothing)
smooth_image = filters.gaussian(image, sigma=1)

# Apply Sobel edge detection filter
edge_image = filters.sobel(image)

# Display original and filtered images
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))

ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(smooth_image, cmap='gray')
ax[1].set_title('Smoothed Image')
ax[1].axis('off')

ax[2].imshow(edge_image, cmap='gray')
ax[2].set_title('Edge Image')
ax[2].axis('off')

plt.show()
```

## Morphological

### Built-in - Medical

```python
from skimage import morphology, data
import matplotlib.pyplot as plt

# Load built-in cell image
image = data.cell()

# Apply morphological erosion
eroded_image = morphology.erosion(image)

# Apply morphological dilation
dilated_image = morphology.dilation(image)

# Display the original and filtered images
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))

ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(eroded_image, cmap='gray')
ax[1].set_title('Eroded Image')
ax[1].axis('off')

ax[2].imshow(dilated_image, cmap='gray')
ax[2].set_title('Dilated Image')
ax[2].axis('off')

plt.show()
```
